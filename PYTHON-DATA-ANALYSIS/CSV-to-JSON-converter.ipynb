```python
import pandas as pd
import numpy as np
import json
```

```python
fileName="./data.csv"
myDict={}
df = pd.read_csv(fileName, encoding='euc-kr')
for i in df.values:
    adress = i[5]
    if adress in myDict :
        myDict[adress][3].append(i[4])
        myDict[adress][4]=len(myDict[adress][3])
    else:
        myDict[adress] = [i[5], i[7], i[8], [i[4]], 1]
myDict['제주특별자치도 서귀포시 남원읍 신례리']
for i in myDict.values():
    print(i)
    
filePath = "flooding_data.json"

dictList=[]
for i in myDict.values():
    dictList.append({"adress": i[0], "latitude": i[1], "longitude": i[2], "contents": i[3], "quantity": i[4]})
with open(filePath, 'w', encoding='utf-8') as outfile:
    json.dump(dictList, outfile, indent="\t", ensure_ascii=False)
```
